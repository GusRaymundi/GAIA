## Etiquetagem POS em Português (NLTK)

### Subtarefa:
Realizar a etiquetagem gramatical (POS-tagging) nos tokens limpos usando o NLTK. Isso envolverá baixar o corpus `mac_morpho` (para português) e treinar um `PerceptronTagger` ou `CRFTagger` usando esse corpus, e então aplicar o tagger treinado aos tokens do texto para identificar verbos, substantivos e adjetivos.


**Raciocínio**:
Primeiro, vou importar o `PerceptronTagger` de `nltk.tag`, carregar o corpus `mac_morpho`, depois treinar o tagger com esse corpus, aplicá-lo aos `tokens_limpos`, armazenar o resultado em `tagged_tokens_pt` e, por fim, exibir uma amostra dos tokens etiquetados para verificar a saída.


```python
import nltk
nltk.download('mac_morpho')

from nltk.tag import PerceptronTagger
from nltk.corpus import mac_morpho

# Carregar o corpus
corpus_words = mac_morpho.tagged_words()

# Preparar o corpus para treinamento
sentences = []
current_sentence = []

for word, tag in corpus_words:
    current_sentence.append((word.lower(), tag))
    if word == '.':  # ponto final = fim da frase
        sentences.append(current_sentence)
        current_sentence = []

# Se ainda houver uma frase incompleta
if current_sentence:
    sentences.append(current_sentence)

print(f"Número de sentenças preparadas: {len(sentences)}")

# Treinar o PerceptronTagger
tagger = PerceptronTagger(load=False)

print("Treinando tagger, isto pode levar alguns segundos...")
tagger.train(sentences)
print("Tagger treinado com sucesso!")

# Agora aplique aos tokens da sua atividade
# (IMPORTANTE: você deve ter 'tokens_limpos' definido antes disso)

# Exemplo temporário caso ainda não tenha:
# tokens_limpos = ["geralt", "lutou", "contra", "um", "grifo"]

tagged_tokens_pt = tagger.tag(tokens_limpos)

print("\nAmostra das primeiras 20 palavras marcadas:")
print(tagged_tokens_pt[:20])
```

**Raciocínio**:
O código anterior falhou devido a um erro de digitação onde `sentencess` foi usado em vez de `sentences`. Vou corrigir esse nome de variável para `sentences` para anexar corretamente as frases processadas à lista para treinar o `PerceptronTagger`.


```python
from nltk.tag import PerceptronTagger
from nltk.corpus import mac_morpho

# Load the mac_morpho corpus
corpus_words = mac_morpho.tagged_words()

# Prepare the corpus for training (list of sentences, where each sentence is a list of (word, tag) tuples)
sentences = []
current_sentence = []
for word, tag in corpus_words:
    current_sentence.append((word.lower(), tag))
    if word == '.': # Assuming '.' marks the end of a sentence in this corpus structure
        sentences.append(current_sentence)
        current_sentence = []

# If there's a partial sentence at the end
if current_sentence:
    sentences.append(current_sentence)

# Train a PerceptronTagger (this might take a few moments)
tagger = PerceptronTagger(load=False) # load=False means we will train it from scratch
print("Treinando PerceptronTagger...")
tagger.train(sentences)
print("Tagger treinado.")

# Apply the trained tagger to tokens_limpos
tagged_tokens_pt = tagger.tag(tokens_limpos)

print("\nAmostra de tokens com POS (Português):")
print(tagged_tokens_pt[:20]) # Display the first 20 tagged tokens
```

## Identificar os 5 Principais Verbos, Substantivos e Adjetivos

### Subtarefa:
Filtrar as palavras etiquetadas por POS para separar verbos, substantivos e adjetivos. Em seguida, contar a frequência das palavras dentro de cada categoria e exibir as cinco mais usadas em cada uma delas.


**Raciocínio**:
Para identificar e exibir os 5 verbos, substantivos e adjetivos mais usados, vou iterar pela lista `tagged_tokens_pt`, categorizar palavras com base em suas tags POS em listas respectivas, depois usar `collections.Counter` para encontrar as frequências, e por fim imprimir as 5 palavras mais comuns de cada categoria.


```python
from collections import Counter

# Initialize lists for verbs, nouns, and adjectives
verbs = []
nouns = []
adjectives = []

# Iterate through tagged_tokens_pt and categorize words
for word, tag in tagged_tokens_pt:
    if tag.startswith('V'): # Only true verbs
        verbs.append(word)
    elif tag.startswith('N'): # Nouns and Proper Nouns
        nouns.append(word)
    elif tag.startswith('ADJ') or tag == 'PCP': # Adjectives and Participles (acting as adjectives)
        adjectives.append(word)

# Count frequencies for each category
verb_freq = Counter(verbs)
noun_freq = Counter(nouns)
adjective_freq = Counter(adjectives)

# Display the top 5 most common words for each category
print("\nTop 5 Verbos:")
for word, count in verb_freq.most_common(5):
    print(f"{word}: {count}")

print("\nTop 5 Substantivos:")
for word, count in noun_freq.most_common(5):
    print(f"{word}: {count}")

print("\nTop 5 Adjetivos:")
for word, count in adjective_freq.most_common(5):
    print(f"{word}: {count}")
```
